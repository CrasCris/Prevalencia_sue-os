```python
import pandas as pd
import numpy as np



def load_data():
    
    df_np = pd.read_excel("base_pediatria.xlsx", sheet_name="BASE DE DATOS OF-NIÑOS DE PAPEL", engine="openpyxl")

    df_sl = pd.read_excel("base_pediatria.xlsx", sheet_name="BASE DE DATOS OF SAN LUIS ", engine="openpyxl")


    return df_np, df_sl

df_np, df_sl = load_data()


# Concatenate both dataframes
df = pd.concat([df_np, df_sl], ignore_index=True)   
```

```python
# Socio-demografico

import matplotlib.pyplot as plt
import seaborn as sns

# supongamos df tiene columnas 'x' y 'y'
# 1) matplotlib (slice por posición)
sc = df['Sociodemográfica'].iloc[1:]
plt.plot(sc)  # omitir la primera fila    
plt.title("Omitiendo primera fila (matplotlib)")
plt.xlabel("x"); plt.ylabel("y")
plt.show()
```
